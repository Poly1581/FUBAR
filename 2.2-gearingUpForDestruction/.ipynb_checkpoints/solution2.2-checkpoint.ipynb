```python
def solution(pegs):
    dists = [pegs[i+1]-pegs[i] for i in range(len(pegs)-1)]
    altSum = 0
    for i,dist in enumerate(dists[::-1]):
        if i % 2 == 0:
            print(f"\t\tadding {dist} to altsum")
            altSum += dist
        else:
            print(f"\t\tsubtracting {dist} from altsum")
            altSum -= dist
    print(f"\taltSum: {altSum}")
    radii = [0]
    if len(pegs) % 2 == 0:
        print(f"\teven, ri: {2*altSum}")
        radii[0] = 2*altSum
        dists = [d*3 for d in dists]
        print(f"\t{dist}")
    else:
        print(f"\todd, ri: {-2*altSum}")
        radii[0] = -2*altSum
    while len(radii) < len(pegs) and radii[-1] > 0:
        print(f"\t{dists[len(radii)-1]} - {radii[-1]}")
        radii.append(dists[len(radii)-1] - radii[-1])
        print(f"\tprevRad: {radii[-1]}")
    print(f"\tradii:{radii}")
    if radii[-1] <= 0:
        return [-1,-1]
    else:
        return [radii[0], 3 if len(pegs) % 2 == 0 else 1]
```

```python
#TEST CASES:
inputs = [[4, 17, 50], [4, 30, 50], [3, 30, 40, 60], [0, 5],[0,5,10]]
expected = [[-1,-1], [12,1], [74,3], [10, 3],[-1,-1]]
for (i,e) in zip(inputs, expected):
    print(f"TESTING {i}")
    a = solution(i)
    print(f"\tsolution({i}) = {a}")
    if e == a:
        print("\tTEST PASSED")
    else:
        print(f"\tTEST FAILED, EXPECTED {e}")
```
