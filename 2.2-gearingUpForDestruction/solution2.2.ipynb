```python
def solution(pegs):
    dists = [pegs[i+1]-pegs[i] for i in range(len(pegs)-1)]
    altSum = 0
    for i,dist in enumerate(dists[::-1]):
        if i % 2 == 0:
            print(f"\t\tadding {dist} to altsum")
            altSum += dist
        else:
            print(f"\t\tsubtracting {dist} from altsum")
            altSum -= dist
    print(f"\taltSum: {altSum}")
    radii = [0]
    minRad = 1
    if len(pegs) % 2 == 0:
        print(f"\teven, ri: {2*altSum}")
        radii[0] = 2*altSum
        dists = [d*3 for d in dists]
        minRad = 3
        print(f"\tdist: {dist}")
    else:
        print(f"\todd, ri: {-2*altSum}")
        if altSum >= 0:
            return [-1,-1]
        radii[0] = -2*altSum
    while len(radii) < len(pegs) and radii[-1] >= minRad:
        print(f"\t{dists[len(radii)-1]} - {radii[-1]}")
        radii.append(dists[len(radii)-1] - radii[-1])
        print(f"\tprevRad: {radii[-1]}")
    print(f"\tradii:{radii}")
    if radii[-1] < minRad:
        return [-1,-1]
    else:
        num = 1
        if len(pegs) % 2 == 0:
            if radii[0] % 3 == 0:
                radii[0] = radii[0]//3
            else:
                return [radii[0], 3]
        return [radii[0], 1]
        # return [radii[0], 3 if len(pegs) % 2 == 0 else 1]
```
